```python
from nltk.corpus import stopwords
stop_words = stopwords.words('indonesian')
print(stop_words)
```

```python
stop_words.extend(["iyahhh","xixixi"])
```

```python
print(stop_words)
```

```python
#stemming
#from nltk.stem import PorterStemmer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
input_str = "ceo formula alberto longgo menampik bhw jakarta membayar kota ternama dunia men"
input_str = nltk.word_tokenize(input_str)
for word in input_str:
    print(stemmer.stem(word))
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
input_str = "terkait isu pemilihan vanue formula ditentukan presiden tegaskan tanggung"
input_str = nltk.word_tokenize(input_str)
for word in input_str:
    print(lemmatizer.lemmatize(word))
```

```python
data['STOP_REMOVAL'][0:10]
```

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()
bebas = []
for index,row in data.iterrows():
    filter_sentence = []
    sentence = row['STOP_REMOVAL']
    words = nltk.word_tokenize(sentence)
    words = [w for w in words if not w in stop_words]
    for word in words:
        filter_sentence.append(stemmer.stem(word))
    bebas.append(filter_sentence)
        
#print(bebas)
#data = data.append(dict(zip(data.columns, filter_sentence)), ignore_index=True)
data["stemming"] = bebas
print(data.columns)
```

```python
def fit_stopwords(text):
    text = np.array(text)
    text = ' '.join(text)
    return text

data['STEMMING'] = data['stemming'].apply(lambda x: fit_stopwords(x))
data.head(10)
```
